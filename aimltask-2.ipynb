```python
import pandas as pd

file_path = "C:/Users/ajith/Downloads/extracted_dataset/Titanic-Dataset.csv"
df = pd.read_csv(file_path)

print(df.info())  # Check data structure
print(df.head())  # Preview first few rows
```

```python
print(df.describe())  
print(df.isnull().sum())  
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram for Age
sns.histplot(df["Age"].dropna(), bins=30, kde=True)
plt.title("Age Distribution")
plt.show()

# Boxplot for Fare
sns.boxplot(x=df["Fare"])
plt.title("Fare Distribution")
plt.show()
```

```python
sns.pairplot(df[["Survived", "Age", "Fare", "Pclass"]], hue="Survived")
plt.show()

# Correlation matrix
import seaborn as sns
import numpy as np

plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Feature Correlation Matrix")
plt.show()
```

```python
print(df.groupby("Sex")["Survived"].mean())  # Survival rate by gender
print(df.groupby("Pclass")["Survived"].mean())  # Survival rate by class
```
